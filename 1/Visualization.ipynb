```python
import pandas as pd
import seaborn as sns
#import matplotlib.pyplot as plt
import numpy as np
```

Objective: Given a flower, can we determine if it belongs to virginica, setosa, versicolor?

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
# put iris features and response into numpy n-dimensional array --> dataframe
# name the columns appropriately, removing the " (cm)" part
X = pd.DataFrame(iris.data, columns=[x.replace(' (cm)','') for x in iris.feature_names])
X['target'] = pd.Series(iris.target)
Y =  pd.Series(iris.target)
```

## Plot the disrbution in a FacetGrid by target/or some other categorical data  
This can tell us further, that petal length is potentially an even better feature to distinguish target 0 from the other two.  
And also a reasonably good separator between target 1 and target 2.

```python
sns.set_style("white")
g = sns.FacetGrid(X, row="target", height=1.7, aspect=4)
g = g.map(sns.distplot, "petal length").add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
g = sns.FacetGrid(X, hue="target", aspect=1.7, size=5)
g = g.map(plt.scatter, "petal length", "petal width").add_legend()
plt.show()
```
